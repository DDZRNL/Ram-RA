```python
from bs4 import BeautifulSoup
import os
import pandas as pd
import csv
import json
import re
```

```python
def ReadFiles(Directory_path):     # Read all the html files
    path=Directory_path   # The directory
    file_names = os.listdir(path)  # 得到文件夹下的所有文件名称
    files_position=[]
    for file_name in file_names:
        position=path+"/"+file_name
        #print(position)
        files_position.append(position)

    print("Total number of HTML files: ", len(files_position))
    return files_position
```

```python
### Merge File
exportFileName = './output/task1_SBIR_all.csv'
files = ReadFiles("./Task1Result")
files.sort()
flag = True
for fileName in files:
    print(fileName)
    if flag:
        pd.read_csv(fileName, encoding="ISO-8859-1").to_csv(exportFileName,index=False, encoding="utf-8")
        flag = False
    else:
        pd.read_csv(fileName, encoding="ISO-8859-1").to_csv(exportFileName,index=False, encoding="utf-8", header=False, mode='a')
```
