```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Data with detail information

```python
def to_CSV(totalData, fileName):
    pd.DataFrame(totalData).to_csv(fileName,index=False, encoding="utf-8")
```

```python
### Read Data
dataDetail = pd.read_csv("task8-V4.csv")
dataDetail.head(50)
```

```python
onehotDetail = pd.get_dummies(dataDetail, columns=['Agency'])
onehotDetail.head()
```

```python
dataDetail["Agency"].value_counts()
```

```python
### Agency Overview
index = dataDetail["Agency"].value_counts().index
onehotGroup = onehotDetail.groupby(['Company'])
for postfix in index:
    realIndex = 'Agency_' + postfix
    sumData = onehotGroup[realIndex].sum()
    print("Awarded number by company and agency: " + postfix)
    print(sumData.describe())
```

### Original Data with no detail information

```python
data2 = pd.read_csv("data.csv")
```

```python
nDF = pd.DataFrame(data2, columns=["Company Name", "Experience With SBIR Program"])
sortedDF = nDF.sort_values(by=["Experience With SBIR Program"], ascending=False)
```

```python
### Mean Value
print("Mean of award Number: " + str(nDF["Experience With SBIR Program"].mean(axis=0)))
### STD
print("Standard deviation of award Number: " + str(nDF["Experience With SBIR Program"].std(axis=0)))
### Max
print("Max of award Number: " + str(nDF["Experience With SBIR Program"].max(axis=0)))
### Min
print("Min of award Number: " + str(nDF["Experience With SBIR Program"].min(axis=0)))
### Median
print("Median of award Number: " + str(nDF["Experience With SBIR Program"].median(axis=0)))
### Statistics
print(nDF.describe())
### Box Plot
nDF.boxplot(column="Experience With SBIR Program")
```

```python
### Save
to_CSV(sortedDF, "counts.csv")
```
